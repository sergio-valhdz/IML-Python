# Algoritmo de k-medias

Para aprendizaje no supervisado, que resuelve problemas de clusterización. Parte un conjunto de 'n' observaciones en 'k' grupos, en el que cada observacion pertenece al grupo cuyo valor medio es más cercano.
Se utiliza en minería de datos

## Ejemplo

```python
import pandas as pd 
import numpy as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import make_blobs
```

```python
datos = make_blobs(n_samples=200, n_features=2, centers=4)
```

```python
datos
```

```python
datos[0]
```

```python
datos[1]
```

```python
plt.scatter(datos[0][:,0],datos[0][:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters=4)
```

```python
modelo.fit(datos[0])
```

```python
modelo.cluster_centers_
```

```python
modelo.labels_
```

```python
datos[1]
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=[12,4])

ax1.scatter(datos[0][:,0],datos[0][:,1], c=modelo.labels_)
ax1.set_title('Algoritmo de k-medias')

ax2.scatter(datos[0][:,0],datos[0][:,1], c=datos[1])
ax2.set_title('Datos originales')
```
