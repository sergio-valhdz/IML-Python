```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(2,9,10)
```

```python
y = x**2
```

```python
plt.plot(x,y,'g')
plt.title("Grafico de una ecuacion cuadratica")
plt.xlabel("Mis valores don de numpy")
plt.ylabel("Mis valores son de x al cuadrado")
plt.show()
```

```python
#Podemos ordenar las graficas, el primero es de filas, columna y la posicion
plt.subplot(1,2,1)
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'g')

plt.subplot(2,1,2)
plt.plot(x,y,'b')
```

## Programacin orientada a objetos

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.8,0.8])
grafico.plot(x,y,'green')
```

```python
figura = plt.figure()
grafico_uno = figura.add_axes([0.1,0.1,0.9,0.9])
grafico_uno.plot(x,y,'r')

grafico_dos = figura.add_axes([0.15,0.5,0.4,0.3])
grafico_dos.plot(y,x,'blue')
```

## Multigraficos

```python
figura, graficos = plt.subplots(nrows=4,ncols=3)
```

```python
figura, graficos = plt.subplots(nrows=2,ncols=2)

graficos[0][0].plot(x,y,'b')
graficos[0][1].plot(x,y,'r')
graficos[1][0].plot(x,y,'g')
graficos[1][1].plot(x,y,'y')

graficos[0][0].set_title('Azul')
graficos[0][1].set_title('Rojo')
graficos[1][0].set_title('Vere')
graficos[1][1].set_title('Amarillo')
```

## Tamao de la figura

```python
figura = plt.figure(figsize=(8,4))
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r')
```

```python
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.9,0.9])
graficos.plot(x,x**2,label='Al cuadrado')
graficos.plot(x,x**3,label='Al cubo')
graficos.legend(loc=0)
```

## Color del grafico

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'#0b0264')
grafico.plot(x,x**4,color='red')
```

## Linea de grafico (grosor, transparencia y tipo)

```python
grosorin = plt.figure()
grafiquin = grosorin.add_axes([0,0,1,1])
grafiquin.plot(x,y,color='blue',linewidth=15,alpha=0.3,linestyle='--')
grafiquin.plot(x,y-4,color='blue')
```
