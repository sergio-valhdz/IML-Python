# Máquinas de vectores de soporte (SVM)

Las SVM son un conjunto de algoritmos de aprendizaje supervisado para resolver clasificación y regresión.

Podemos etiquetar clases y usar SVM para construit un modelo que prediga la clase de una muestra

## Ejercicio

```python
import pandas as pd 
import numpy as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
flores = sns.load_dataset('iris')
```

```python
flores.head()
```

```python
flores['species'].unique()
```

```python
sns.pairplot(flores)
```

```python
sns.pairplot(flores,hue='species')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = flores.drop('species',axis=1)
```

```python
y = flores['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
modelo = SVC(gamma='auto')
```

```python
modelo.fit(X_train,y_train)
```

```python
predicciones = modelo.predict(X_test)
```

```python
predicciones
```

```python
y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predicciones))
```

```python
print(confusion_matrix(y_test,predicciones))
```
